```python
task_folder = os.path.dirname(os.path.abspath("__file__"))
task_folder = os.path.join(task_folder, 'Tasks', 'GridMLP_decoupled_resolution_levels_network_sin')
my_mkdir(task_folder)
print(task_folder)
task_hyperparameter = {
	"encoding": {
		"otype": "Grid",
		"type": "Dense",
		"n_levels": 12,
		"n_features_per_level": 1,
		"log2_hashmap_size": 30,
		"base_resolution": 62,
		"per_level_scale": 1.12,
        "interpolation":"Linear"
	},
	"network": {
		"type": "MLP",
		"activation": "Sin",
		"n_neurons": 32,
		"n_hidden_layers": 1,
		"spectral_norm": 1,
	},
    "loss": {
        "type": "energy",
        "lambda": 1.0,
	},
	"optimizer": {
		"type": "ADAM",
		"learning_rate": 0.005,
		"max_iter": 0,
		"n_step_decay":500,
		"gamma":0.4
    },
	"training":{
		"n_steps": 6000,
		"interior_batch": 30000,
		"boundary_batch": 2000,
		"n_step_output":600,
	},
	"pretrain":{
        "n_steps": 500,
        "n_step_output": 100,
        "n_step_decay": 100,
    },
	"experiment":{
		"n_test": 5,
	}
}

per_level_scale_0 = 1.12
n_neurons_list = [16,32,64]
n_hidden_layers_list = [0,1,2]

n_levels_list = np.array([1,4,8,16])
max_res_list = np.array([16,32,64,128,256])

i = 0
for n_levels in n_levels_list:
	base_resolution_list = np.round((max_res_list/(per_level_scale_0**(n_levels-1)))).astype("int")
	print(n_levels,base_resolution_list,(base_resolution_list*per_level_scale_0**(n_levels-1)).astype('int'))
	for base_resolution in base_resolution_list:
		for n_hidden_layers in n_hidden_layers_list:
			for n_neurons in n_neurons_list:
				# if n_hidden_layers == 0 and n_neurons == 16:
				# 	continue
				# if n_hidden_layers == 2 and n_neurons == 32:
				# 	continue
				task_hyperparameter["network"]['n_neurons'] = n_neurons
				task_hyperparameter["network"]['n_hidden_layers'] = n_hidden_layers
				task_hyperparameter["encoding"]['base_resolution'] = float(base_resolution)
				task_hyperparameter["encoding"]["n_levels"] = float(n_levels)
				file_path = os.path.join(task_folder, 'config_{}.json'.format(i))
				with open(file_path, 'w') as f:
					json.dump(task_hyperparameter, f, ensure_ascii=False)
				i += 1
print(i)
```
